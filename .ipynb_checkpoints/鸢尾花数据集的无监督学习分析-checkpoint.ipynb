```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import seaborn as sns
sns.set(style='white',context='notebook',palette='muted')

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
# 读取数据
df_iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',
                      names=['sepal length', 'sepal width', 'petal length', 'petal width', 'class'])
# df_iris
# 处理缺失值
df_iris = df_iris.dropna()

# 处理异常值
df_iris = df_iris[(df_iris['sepal length'] > 0) & (df_iris['sepal width'] > 0)
                  & (df_iris['petal length'] > 0) & (df_iris['petal length'] > 0)]

# 数据可视化
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

ax[0, 0].hist(df_iris['sepal length'], bins=20)
ax[0, 0].set_title('sepal length')

ax[0, 1].hist(df_iris['sepal width'], bins=20)
ax[0, 1].set_title('sepal width')

ax[1, 0].hist(df_iris['petal length'], bins=20)
ax[1, 0].set_title('petal length')

ax[1, 1].hist(df_iris['petal width'], bins=20)
ax[1, 1].set_title('petal width')
```

```python
# df_iris.describe()
df_iris_copy = df_iris
```

```python
lis = df_iris.loc[:, ['class']].values
y =  [0 if x == 'Iris-setosa' else 1 if x == 'Iris-versicolor' else 2 for x in lis]
# lis.shape,X.shape
df_iris_copy['class'] = y
df_iris_copy['petal length'].describe()
```

```python
facet = sns.FacetGrid(df_iris_copy,hue="class",aspect=4)
facet.map(sns.kdeplot,'sepal length',shade=True)
facet.set(xlim=(3,9))
facet.add_legend()
```

```python
facet = sns.FacetGrid(df_iris_copy,hue="class",aspect=4)
facet.map(sns.kdeplot,'petal length',shade=True)
facet.set(xlim=(1,7))
facet.add_legend()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
# 对数据进行切片，获得特征数据
X = df_iris.iloc[:, :-1].values
# 数据预处理：使得每个特征的均值为0，标准差为1
X_std = StandardScaler().fit_transform(X)
# X_std.shape
kmeans = KMeans(n_clusters=3,random_state=42)
y_kmeans = kmeans.fit_predict(X_std)
# y_kmeans
# 可视化聚类结果
plt.scatter(X_std[y_kmeans == 0, 0], X_std[y_kmeans == 0, 1],s=100, c='red', label='Cluster 1')
plt.scatter(X_std[y_kmeans == 1, 0], X_std[y_kmeans == 1, 1],s=100, c='green', label='Cluster 2')
plt.scatter(X_std[y_kmeans == 2, 0], X_std[y_kmeans == 2, 1],s=100, c='blue', label='Cluster3')
# 在散点图中绘制聚类中心点
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')
plt.title('K-means Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
Z = linkage(X_std,method='ward')

# 可视化树状图
plt.figure(figsize=(8,6))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.decomposition import PCA
# 创建PCA对象
pca_three = PCA(n_components=3)
# 对数据集进行PCA降维
principalComponents_three = pca_three.fit_transform(X_std)
# 可视化PCA降维后的数据
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
ax.scatter(principalComponents_three[:, 0], principalComponents_three[:, 1], principalComponents_three[:, 2],
           c=y, cmap=plt.cm.Set1, edgecolor='w', s=40)
ax.set_title('3 Component PCA on Iris Dataset', fontsize=20)

ax.set_xlabel('principal component 1', fontsize=15)
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel('principal component 2', fontsize=15)
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel('principal component 3', fontsize=15)
ax.w_zaxis.set_ticklabels([])
plt.show()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X_std, y)
plt.figure(figsize=(10, 6))
plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y, cmap='viridis')
plt.xlabel('LD1')
plt.ylabel('LD2')
# plt.legend()
plt.title('LDA on Iris Dataset')
plt.show()
```
