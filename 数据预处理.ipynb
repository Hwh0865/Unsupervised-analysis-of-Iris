```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据集
#  Pandas 库中的 read_csv 函数从指定的 URL 地址中读取 CSV 文件，并将其存储在名为 df 的 Pandas DataFrame 对象中。
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",
                 names=['sepal length', 'sepal width', 'petal length', 'petal width', 'class'])

df_14 = df
df_14
```

```python
# 数据可视化
# fig 是一个 Figure 对象，表示整个图形，而 axs 是一个包含 2 行 2 列的 axis 对象的 2 维数组，表示图形网格中的每个子图
# fig, axs = plt.subplots(2, 2, figsize=(10, 10))
# axs
```

```python
# .hist() 是这个对象的一个方法，用于创建一个直方图。
# df['sepal length'] 表示 DataFrame 中的 'sepal length' 这一列数据，bins=20 表示将这些数据分成 20 个区间，
# 并统计每个区间的频数，生成一个直方图，显示 'sepal length' 这一列数据的分布情况。
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].hist(df['sepal length'], bins=20)
axs[0, 0].set_title('Sepal Length')
plt.show()
```

```python
# 处理缺失值
df = df.dropna()

# 处理异常值
df = df[(df['sepal length'] > 0) & (df['sepal width'] > 0) & (df['petal length'] > 0) & (df['petal width'] > 0)]

# 数据可视化
# plt.subplots 是 Matplotlib 库中的一个函数，用于创建包含多个子图的图形。
# 具体来说，plt.subplots 函数可以指定图形的行数和列数，然后返回一个包含整个图形以及每个子图的 axis 对象的元组。
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0, 0].hist(df['sepal length'], bins=20)
axs[0, 0].set_title('Sepal Length')

axs[0, 1].hist(df['sepal width'], bins=20)
axs[0, 1].set_title('Sepal Width')

axs[1, 0].hist(df['petal length'], bins=20)
axs[1, 0].set_title('Petal Length')

axs[1, 1].hist(df['petal width'], bins=20)
axs[1, 1].set_title('Petal Width')

plt.show()
```
